```python
X=df[['Size','Weight','Sweetness','Crunchiness','Juiciness','Ripeness','Acidity']]
Y=df['Quality']
```

```python
from sklearn.preprocessing import StandardScaler
sv=StandardScaler()
X_scalar=sv.fit_transform(X)
X_scalar
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X_scalar,Y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,criterion='gini', oob_score=  True)
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,Y_test)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(penalty='l2',dual=False,C=1.0)
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
accuracy_score(y_pred,Y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(X_train,Y_train)
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
accuracy_score(y_pred,Y_test)
```
